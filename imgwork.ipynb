# read tiled image

```python
filename = "1061781.78x5977083.324816011x1066672.779808x5981974.324624011.png"
```

```python
im = Image.open(f"out/{filename}").getdata()
```

```python
rgb_im = im.convert("RGB")
```

###  find red pixels

```python
match_pixel = (226, 0, 116)
```

```python
rgb_im.size
```

```python
matched_pixels = []
```

```python
for x in range(rgb_im.size[0]):
    for y in range(rgb_im.size[1]):
        r, g, b = rgb_im.getpixel((x, y))
        if (r,g,b) == match_pixel:
            matched_pixels.append((x,y))
            # print(f"Found {match_pixel} at {x},{y}!")
```
